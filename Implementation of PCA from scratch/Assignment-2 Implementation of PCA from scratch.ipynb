```python
import pandas as pd
```

```python
import numpy as np
```

```python
data=pd.read_csv("aps_failure_training_set.csv")
```

```python
data
```

```python
class_labels = []

for values in data['class']:
    if values == 'neg':
        class_labels.append(0)
    else:
        class_labels.append(1)

data['class'] = class_labels
```

```python
data
```

```python
y = data['class']
```

```python
y
```

```python
data.drop(labels=data.columns[0],axis=1,inplace=True)
```

```python
data
```

```python
a = {}
features = list(data)

for i in features:
    column = data[i]
    try:
        nans = column.value_counts()['na']
    except KeyError:
        nans = 0
    a[i] = nans
perc = {f:(c/60000) for f,c in a.items()}
```

```python
a
```

```python
ft = list(perc.values())
```

```python
ft
```

```python
more_than_70_col=[features[i] for i in range(len(ft)) if ft[i]*100>70]
less_than_10_col=[features[i] for i in range(len(ft)) if ft[i]*100<=5]
```

```python
more_than_70_col
```

```python
less_than_10_col
```

```python
data.drop(columns=more_than_70_col,inplace=True)
```

```python
data.replace('na',np.nan,inplace=True)
```

```python
for i in list(data.columns)[1:]:
    data[i]=data[i].astype(float)
```

```python
data.dropna(subset=less_than_10_col,inplace=True)
```

```python
data = data.reset_index(drop=True)
```

```python
data.fillna(data.mean(numeric_only=True).round(1),inplace=True)
```

```python
data
```

```python
x=data
```

```python
x
```

```python
x_mean = x - x.mean()
```

```python
x_mean = np.array(x_mean)
```

```python
x_mean
```

```python
sigma = (1/x.shape[0])*np.matmul(x_mean.T,x_mean)
```

```python
sigma
```

```python
sigma_cov = x.cov()
```

```python
sigma_cov
```

```python
factorized_matrices = np.linalg.svd(sigma)
```

```python
factorized_matrices
```

```python
projected_in_new_space = np.matmul(x_mean,factorized_matrices[0])
```

```python
projected_in_new_space
```

```python
projected_in_new_space.shape
```

```python
projected_in_new_space = pd.DataFrame(data=projected_in_new_space,columns=["Feature_"+str(i) for i in range(1,164)])
```
